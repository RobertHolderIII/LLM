## download Jammer's homepage for DS9 reviews

```python
import re
from requests import get

# pull all the PHP links from the home page 
home = 'https://www.jammersreviews.com/st-ds9/'
response = get(home)
```

## extract all the review links from the homepage
Had to do make the regex pretty specific to avoid picking up stray matches

```python
pattern = r'''/st-ds9
              /s.         # season e.g. s1, s2
              /\w*        # title of the review
              \.php       # .php extension
            '''
php_regex = re.compile(pattern, re.VERBOSE)
review_urls = php_regex.findall(response.text)

review_urls
```

```python
import pandas as pd
from bs4 import BeautifulSoup

reviews = []

prefix = 'https://www.jammersreviews.com'
for idx,url in enumerate([prefix + suffix for suffix in review_urls]):
    #download review
    print(f'{idx+1}/{len(review_urls)}: downloading {url}...')
    html = get(url).text
    
    
    #TODO separate the review and the comment sections
    review = html
    comments = ""
    
    #extract text and store in pandas
    reviews.append({
        'url': url,
        'review': BeautifulSoup(review, features="html.parser").get_text().strip(),
        'comments': BeautifulSoup(comments, features="html.parser").get_text().strip()
        })

reviews_df = pd.DataFrame.from_dict(reviews)
reviews_df
```
